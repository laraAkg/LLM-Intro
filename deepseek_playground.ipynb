# Example to Run Deepseek locally / in Google colab with ollama

Prerequisites: Nvidia GPU with Cuda, Unix/Linux environment -> DESHALB ZWINGEND IM COLAB ÖFFNEN


## Download ollama and install ollama

```python
!curl https://ollama.ai/install.sh | sh
```

## Install cuda drivers
- The first command disables the interactive shell, as we do not have access to it in colab
- the second command installs the latest version of the cuda drivers to optimize the inference (predicition) of the model

```python
!echo 'debconf debconf/frontend select Noninteractive' | sudo debconf-set-selections
!sudo apt-get update && sudo apt-get install -y cuda-drivers
```

## Ensure the cuda drivers are used

```python
import os

# Set LD_LIBRARY_PATH so the system NVIDIA library
os.environ.update({'LD_LIBRARY_PATH': '/usr/lib64-nvidia'})
```

## Start Ollama
- nohup is a command line expression that prevents a process from stopping after exiting the terminal.

```python
!nohup ollama serve &
```

# Donwload the deepseek model

```python
!ollama pull deepseek-r1:7b
```

## Install ollama in your Python environment

```python
!pip install ollama
```

## Example model request

```python
import ollama
response = ollama.chat(model='deepseek-r1:7b', messages=[
  {
    'role': 'user',
    'content': "How many r's are in a strawberry?",
  },
])
print(response['message']['content'])
```

Der Code installiert und verwendet das Deepseek-Modell mit Ollama in Google Colab.

Hier sind die Schritte im Detail:

**1. Ollama herunterladen und installieren:**

*!curl https://ollama.ai/install.sh | sh:* Dieser Befehl lädt das Ollama-Installationsskript herunter und führt es aus, um Ollama zu installieren.

**2. CUDA-Treiber installieren:**

- *!echo 'debconf debconf/frontend select Noninteractive' | sudo debconf-set-selections:* Dieser Befehl deaktiviert die interaktive Shell für die Installation der CUDA-Treiber.

- *!sudo apt-get update && sudo apt-get install -y cuda-drivers:*
Dieser Befehl aktualisiert die Paketliste und installiert die CUDA-Treiber, die für die GPU-Beschleunigung erforderlich sind.

**3. Sicherstellen, dass die CUDA-Treiber verwendet werden:**
*os.environ.update({'LD_LIBRARY_PATH': '/usr/lib64-nvidia'}):* Dieser Code legt die Umgebungsvariable *LD_LIBRARY_PATH* fest, um sicherzustellen, dass das System die NVIDIA-Bibliotheken verwendet.

**4. Ollama starten:**
*!nohup ollama serve &*: Dieser Befehl startet den Ollama-Server im Hintergrund. nohup stellt sicher, dass der Server auch nach dem Schließen des Terminals weiterläuft.

**5. Das Deepseek-Modell herunterladen:**
*!ollama pull deepseek-r1:7b*: Dieser Befehl lädt das Deepseek-Modell von Ollama herunter.

**6. Ollama in der Python-Umgebung installieren:**
*!pip install ollama:* Dieser Befehl installiert die Ollama-Bibliothek in deiner Python-Umgebung, damit du mit Ollama über Python interagieren kannst.

**7. Beispiel für eine Modellanfrage:**
- Dieser Code verwendet die ollama.chat()-Funktion, um eine Anfrage an das Deepseek-Modell zu senden.
